```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
#print(city_data.head())
ride_data = pd.read_csv(ride_data_to_load)
#print(ride_data.head())

# Combine the data into a single dataset
merge_table = pd.merge(ride_data, city_data, on="city")

# Display the data table for preview
merge_table
```

```python
# Split up our data into groups based upon 'city'
city_group = merge_table.groupby("city")

city_df = city_group.first()
# Find out how many rides and average fare in each city
city_df["Number of Rides"] = city_group["ride_id"].count()
city_df["Average Fare"] = city_group["fare"].mean()
city_df
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_types = ["Urban", "Suburban", "Rural"]
urban_x = city_df.loc[city_df["type"] == "Urban", "Number of Rides"]
#print(urban_x.head())
urban_y = city_df.loc[city_df["type"] == "Urban", "Average Fare"]
#print(urban_y.head())
suburb_x = city_df.loc[city_df["type"] == "Suburban", "Number of Rides"]
suburb_y = city_df.loc[city_df["type"] == "Suburban", "Average Fare"]
rural_x = city_df.loc[city_df["type"] == "Rural", "Number of Rides"]
rural_y = city_df.loc[city_df["type"] == "Rural", "Average Fare"]
urban_drivers = city_df.loc[city_df["type"] == "Urban", "driver_count"]
suburb_drivers = city_df.loc[city_df["type"] == "Suburban", "driver_count"]
rural_drivers = city_df.loc[city_df["type"] == "Rural", "driver_count"]

# Build the scatter plots for each city type
urban_plot = plt.scatter(urban_x, urban_y, marker="o", facecolors="lightcoral", edgecolors="black",
            s=10*urban_drivers, alpha=0.5)
suburban_plot = plt.scatter(suburb_x, suburb_y, marker="o", facecolors="lightskyblue", edgecolors="black",
           s=10*suburb_drivers, alpha=0.5)
rural_plot = plt.scatter(rural_x, rural_y, marker="o", facecolors="gold", edgecolors="black",
           s=10*rural_drivers, alpha=0.5)

# Incorporate the other graph properties
plt.xlim(0, 45)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare (Dollars)")
plt.tight_layout()

# Create a legend
plt.legend((urban_plot, suburban_plot, rural_plot),
           ('Urban', 'Suburban', 'Rural'))

# Incorporate a text label regarding circle size
plt.text(25,35,'Note:\nCircle size correlates to\ndrivers per city')

# Save Figure
import os
strFile = "Images/PyberRideSharingData.png"
plt.savefig(strFile)
```

```python
# Show plot
plt.show()
```
